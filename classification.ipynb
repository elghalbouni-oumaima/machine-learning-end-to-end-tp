```python
import tensorflow as tf
import matplotlib.pyplot as plt

# 1. Chargement direct depuis Keras
fashion_mnist = tf.keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# 2. Vérification des dimensions
print(f"Images d'entraînement : {X_train.shape}")  # Doit afficher (60000, 28, 28)
print(f"Images de test : {X_test.shape}")          # Doit afficher (10000, 28, 28)

# 3. Liste des noms de classes (pour l'affichage)
class_names = ["T-shirt/top", "Pantalon", "Pull", "Robe", "Manteau",
               "Sandale", "Chemise", "Basket", "Sac", "Bottine"]

# 4. Afficher la première image pour vérifier
plt.imshow(X_train[0], cmap="binary")
plt.title(class_names[y_train[0]])
plt.axis('off')
plt.show()
```
